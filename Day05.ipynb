```kotlin
fun List<String>.parseInput(): Pair<List<LongRange>, List<Long>> {
    val ranges = takeWhile { it.isNotEmpty() }.map {
        val (start, end) = it.split("-")
        start.toLong()..end.toLong()
    }

    val ids = takeLastWhile { it.isNotEmpty() }.map {
        it.toLong()
    }

    return ranges to ids
}
```

```kotlin
val (ranges, ids) = testInput.lines().parseInput()

ids.count { id -> ranges.any { range -> id in range } }
```

```kotlin
fun List<LongRange>.mergeOverlapping(): List<LongRange> {
    val merged = mutableListOf<LongRange>()

    this.sortedBy { it.first }.forEach { range ->
        val last = merged.removeLastOrNull()
        if (last != null) {
            if (last.endInclusive > range.start) {
                merged.add(last.start..max(last.endInclusive, range.endInclusive))
            } else {
                merged.add(last)
                merged.add(range)
            }
        } else
            merged.add(range)
    }
    return merged.toList()
}
```
