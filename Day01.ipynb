```kotlin
%use adventOfCode
```

```kotlin
val aoc = AocClient.fromEnv().interactiveDay(2025, 1)
```

```kotlin
val input = aoc.input()
```

```kotlin
val input = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82"""
```

```kotlin
val steps = input.lines().map { it.substring(0,1) to it.substring(1).toInt() }
```

```kotlin
var dial = 50
var zeroCount = 0

for ((direction, amount) in steps) {
    if (direction == "R") {
        dial = (dial + amount) % 100
    } else if (direction == "L") {
        dial = (dial + 100 - amount) % 100
    }
    if (dial == 0) {
        zeroCount += 1
    }
}

zeroCount
```

```kotlin
aoc.submitPartOne(zeroCount)
```

```kotlin
var dial = 50
var zeroCount = 0

for ((direction, amount) in steps) {
    val rotation = Math.floorMod(amount, 100)
    val clicks = Math.floorDiv(amount, 100)

    zeroCount += clicks

    if (direction == "R") {
        if (dial + rotation >= 100) {
            zeroCount++
        }
        dial = (dial + rotation) % 100
    } else {

        if (dial != 0 && dial - rotation <= 0) {
            zeroCount++
        }
        dial = Math.floorMod(dial - rotation, 100)
    }
}

zeroCount
```

```kotlin
aoc.submitPartTwo(zeroCount)
```
