```kotlin
%use adventOfCode
```

```kotlin
val aoc = AocClient.fromEnv().interactiveDay(2025, 2)
```

```kotlin
aoc.viewPartOne()
```

```kotlin
val input = aoc.input()
```

```kotlin
val ranges = input.split(",").map {
    val parts = it.split("-")
    (parts[0].toLong() .. parts[1].toLong())
}
```

```kotlin
fun LongRange.invalidIds1(): List<Long> = filter {
    val str = it.toString()

    str.substring(str.length/2) == str.substring(0, str.length/2)
}
```

```kotlin
val part1 = ranges.map { it.invalidIds1().sum() }.sum()
```

```kotlin
aoc.submitPartOne(part1)
```

```kotlin
aoc.viewPartTwo()
```

```kotlin
fun LongRange.invalidIds2(): List<Long> = filter {
    val str = it.toString()
    for (i in 1..str.length / 2) {
        val regex = Regex("(${str.substring(0,i)})+")
        if(str.matches(regex)) {
          return@filter  true
        }
    }

    false
}
```

```kotlin
ranges.map{it.invalidIds2().sum()}.sum().also { aoc.submitPartTwo(it) }
```
